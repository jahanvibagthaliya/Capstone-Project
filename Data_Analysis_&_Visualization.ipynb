# Data Analysis & Data Visualization for DMart Product and Sales Dataset

```python
import pandas as pd
```

```python
dataset = pd.read_csv(r"C:\Users\Vaishnav\Downloads\ML_Project\Clean Dmart sales data.csv")
dataset.head(5)
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset.drop(columns=['Cancellation Date'],inplace=True)
```

```python
dataset.info()
```

# 1. Top 5 Profitable Product Name

```python
dataset['Total Order Value'] = pd.to_numeric(dataset['Total Order Value'],errors='coerce')   
```

```python
dataset['Total Order Value'].dtype
```

```python
dataset.info()
```

```python
dataset['Total Order Value']=dataset['Total Order Value'].fillna(0)   # Fill NA with 0
```

```python
dataset.info()
```

- Now our dataset is clean and there is no null value presentation in our dataset

```python
dataset['Profit'] = dataset['MRP'] - dataset['Discount Price']
```

```python
top_product = dataset.groupby('Product Name')['Profit'].sum().reset_index()
top_product
```

- Above table shows Profit of different Product Name

```python
top_5_profitable_product = top_product.sort_values(by='Profit',ascending=False)
top_5_profitable_product.head(5)
```

- Above table shows list of Top 5 Profitable Product Names

# 2. Visual Representation of Top 5 Profitable Product using Bar Chart

```python
import matplotlib.pyplot as plt
```

```python
top_5_profitable_product = top_5_profitable_product.head(5)
```

```python
plt.figure(figsize=(20,8))
plt.bar(top_5_profitable_product['Product Name'],top_5_profitable_product['Profit'],color = 'olive')
plt.xlabel("Product Name")
plt.ylabel("Profit")
plt.title("Top 5 Product by Profit")
plt.show()
```

# 3. Unit sold by Product name

```python
dataset['Estimated Units Sold'] = (dataset['Total Order Value']/dataset['Discount Price']).round().astype(int)
```

```python
dataset['Estimated Units Sold'].dtype
```

```python
Product_unit_sold = dataset.groupby('Product Name')['Estimated Units Sold'].sum().reset_index()
```

```python
Product_unit_sold = Product_unit_sold.sort_values(by='Estimated Units Sold' ,ascending=False)
Product_unit_sold
```

- From the above table we can say that:
    - Maximum unit sold for "Parle Nutricrunch Digestive Cookies" and it's 1166
    - Minimum unit sold for "Amul Taaza Toned Milk" and it's 1008


# 4. Visual Representation of Unit wise top 7 Selling Product using Pie Chart

```python
import  matplotlib.pyplot as plt
```

```python
top_7 = Product_unit_sold.head(7)
top_7
```

```python
plt.figure(figsize=(5,5))
plt.pie(top_7['Estimated Units Sold'],labels=top_7['Product Name'],autopct='%1.1f%%',shadow=False)
plt.title("Top 7 selling product by Estimated Unit Sold")
plt.axis('equal')
plt.show()
```

# 5. Total Revenue by DMart

```python
Total_Revenue = dataset['Total Order Value'].sum()
print("Total revenue generated by DMart is: ",Total_Revenue)
```

# 6. Revenue generated by different Marketing and Advertisement

```python
Revenue_by_Advertisement = dataset.groupby('Marketing/Advertisement')['Total Order Value'].sum().reset_index()
Revenue_by_Advertisement = Revenue_by_Advertisement.sort_values(by='Total Order Value', ascending=False)
Revenue_by_Advertisement
```

# 7. Visual representation of Total revenue generated by Marketing / Advertisement using Pie Chart

```python
plt.figure(figsize=(5,5))
plt.pie(Revenue_by_Advertisement['Total Order Value'],labels=Revenue_by_Advertisement['Marketing/Advertisement'],autopct='%1.1f%%',shadow=False)
plt.title("Revenue generated by Marketing/Advertisement")
plt.axis('equal')
plt.show()
```

- From the above representation we can conclude that:
    - Maximum revenue generated from Instagram and it's 34.0%
    - Minimum revenue generated from TV and it's 8.2%

# 8. Month wise unit sold in 2023

```python
data_2023 = dataset[dataset['Year']==2023]
data_2023.head(3)
```

```python
Unit_sold_by_month = data_2023.groupby('Month')['Estimated Units Sold'].sum().reset_index()
Unit_sold_by_month = Unit_sold_by_month.sort_values(by='Estimated Units Sold', ascending=False)
Unit_sold_by_month
```

# 9. Visual Representation of Unit sold in 2023 using Line Graph

```python
from matplotlib.markers import MarkerStyle
```

```python
plt.figure(figsize=(12,5))
plt.plot(Unit_sold_by_month['Month'],Unit_sold_by_month['Estimated Units Sold'],color='orange',marker = 'D',markersize = 5 ,mec = 'r')
plt.xlabel("Month")
plt.ylabel("Estimated Units Sold")
plt.title("Unit Sold in 2023")
plt.show()
```

- From the above representation we can conclude that:
    - In November month maximum(695) unit sold for year 2023
    - In June month minimum(594) unit sold for year 2023

# 10. Subscripction wise total order

```python
Subscription_wise_order = dataset.groupby('Subscription')['Order ID'].count().reset_index()
Subscription_wise_order = Subscription_wise_order.sort_values(by='Order ID',ascending=False)
Subscription_wise_order
```

# 11. Visual Representation of Subscripction wise total order using Bar Chart

```python
plt.figure(figsize=(5,5))
plt.bar(Subscription_wise_order['Subscription'],Subscription_wise_order['Order ID'],color='purple')
plt.xlabel("Subscripction")
plt.ylabel("Total Order")
plt.title("Subscripction wise Total Order")
plt.show()
```

- From the above representation we can conclude that:
    - Maximum total order came from "Freepass" customers
    - Minimum total order came from "Premium Plus" customer

# 12. Average Rating of different Product 

```python
Average_Product_Rating = dataset.groupby('Product Name')['Rating'].mean().reset_index()
Average_Product_Rating = Average_Product_Rating.sort_values(by='Rating',ascending=False)
Average_Product_Rating
```

- From the above table we can conclude that:
    - Maximum average rating is for "Amul Taaza Toned Milk" and it's 3.05
    - Minimum average rating is for "VSR Channa Dal" and it's 2.96
